```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
init_printing(use_latex=True)
```

```python
a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,x,y,z = symbols('a b c d e f g h i j k l m n o x y z', real=True)
```

```python
#Punto 4
def prob(t,w,a=1, wfi = 5):
    return a * np.power(np.sin((wfi-w)*t/2) / (wfi-w),2)
```

```python
t = np.linspace(0,5,500)
w = np.linspace(0,10,100)
```

```python
plt.plot(t,prob(t,w=2.5))
plt.title(r'$P_{if} (t)$ con $w = 2.5$ y $w_{fi} = 5$',fontsize=18)
```

```python
plt.plot(w,prob(t=1,w=w),label ='P_if(w)')
plt.title(r'$P_{if} (w)$ con $t = 1$ y $w_{fi} = 5$',fontsize=18)
plt.scatter(np.array([5]*15),np.linspace(0,np.max(prob(t=1,w=w)),15),marker ='+',label='$w_{fi}$ = 5')
plt.legend()
```
