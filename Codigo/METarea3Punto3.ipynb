# Cálculos del punto 3

```python
from sympy import *
import fractions
import numpy as np
init_printing(use_latex=True)
```

```python
Q =Matrix([[0,1/2,0.5],[0,0,1],[3/4,1/4,0]])
Q #Matriz de transición
```

```python
Q.eigenvals() #Autovalores de la matriz. El primer número es el autovalor, el segundo la multiplicidad geométrica.
```

```python
Q.eigenvects() #El primer número es el autovalor, el segundo la multiplicidad algebráica, el tercero el autovector
```

```python
psi=[list(tup[2][0]) for tup in Q.eigenvects()]

psi
```

```python
chi=[list(tup[2][0]) for tup in Q.T.eigenvects()]
Q.T.eigenvects()#Igual que el anterior pero los autovectores ahora son por la izquierda.
```

Se normaliza el autovector de $\lambda = 1$ pues &lt; $\psi_i$ | $\chi_j$ &gt; = $\delta_{ij}$

```python
norma = [sum(chi[0])]
np.array(chi[0])/norma[0] 
```

Se procede a calcular $\lambda_i$ $\chi_i$ &gt; &lt; $\psi_i$ para autovalor.

```python
norma = Matrix(psi[0]).T*Matrix(chi[0])
steady = Matrix(psi[0])*Matrix(chi[0]).T/norma[0]
steady #Note que esta es la matriz de probabilidad del estado estacionario.
```

```python
norma = Matrix(psi[1]).T*Matrix(chi[1])
steady = Matrix(psi[1])*Matrix(chi[1]).T/norma[0]
simplify(steady)
```

```python
norma = Matrix(psi[2]).T*Matrix(chi[2])
steady = Matrix(psi[2])*Matrix(chi[2]).T/norma[0]
simplify(steady)
```
