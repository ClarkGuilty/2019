```python
from sympy import *
import numpy as np
init_printing(use_latex=True)
```

Primero se define la integral a realizar, en este caso será:

```python
t1, t2, t3, x= symbols('t1 t2 t3 x')
R, k, T, L = symbols('R Kb T L')
gamma, delta, t = symbols('\Gamma \Delta t', nonzero=True)
C = cosh(delta*x)-(gamma/delta)*sinh(delta*x)
expre = exp(-gamma*(t1-2*t2+t3))*C.subs(x,t1-t2)*C.subs(x,t2-t3)
#Integral(exp(-t1),t1)
Integral(expre, (t1,0,t), (t2,0,t), (t3,0,t))
```

Y ahora se procede a la integración simbólica usando Sympy. 

```python
integrate(expre, (t1,0,t), (t2,0,t), (t3,0,t))
```
