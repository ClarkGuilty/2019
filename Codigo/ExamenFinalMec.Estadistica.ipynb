```python
#Punto 1
import numpy as np
import matplotlib.pyplot as plt
kb = 1.38e-23
J = 3e-21
def F(x,T):
    return (2*kb*T-3*J)*x**2+(kb*T*x**4) / 12 - kb*T*np.log(2)
```

```python
x = np.linspace(-1-1e-3,1+1e-3,102)
T0 = 0
Tc = 6*J/kb/(1-x**2)
epsilon = Tc*0.95
```

```python
#plt.plot(x,F(x,T0)/J, label = "T = 0")
plt.plot(x[2:-2],F(x[2:-2],Tc[2:-2])/J, label = "T = Tc")
plt.plot(x[2:-2],F(x[2:-2],Tc[2:-2]-epsilon[2:-2])/J, label = "T < Tc")
plt.plot(x[2:-2],F(x[2:-2],Tc[2:-2]+epsilon[2:-2])/J, label = "T > Tc")
plt.legend(prop={'size': 12})
plt.ylim(-8.5,5)
plt.title("Energía libre para distintas temperaturas",fontsize =18)
plt.ylabel("F / N",fontsize = 15)
plt.xlabel("x",fontsize = 15)
plt.savefig('fig1.png', dpi = 600)
```

```python
Tcritico = 6*J/kb
def xeq(T):
    rta = np.copy(T)
    boo = T<Tcritico
    rta[T>Tcritico] = 0
    rta[T<Tcritico] = np.sqrt(3*(6*J/kb/T[T<Tcritico]-1))
    return rta
def xs(T):
    rta = np.copy(T)
    rta[T<Tcritico] = np.sqrt(6*J/kb/T[T<Tcritico]-1)
    rta[T>Tcritico]=0
    return rta
Te = np.linspace(1,Tcritico*2,100)
```

```python
plt.figure()
plt.plot(Te/Tcritico,xs(Te), color = 'red', label = r'$x_s$')
plt.plot(Te/Tcritico,xeq(Te), color = 'blue', label =r'$\pm x_{eq}$')
plt.plot(Te/Tcritico,-xeq(Te), color = 'blue')
#plt.plot(Te/Tcritico,-xs(Te), color = 'red')
plt.legend(prop={'size': 15})
plt.title(r"Frontera $\pm x_{eq}$ y $x_s$  ",fontsize =18)
plt.ylabel("x",fontsize = 15)
plt.xlabel(r"$T/T_c$",fontsize = 15)
plt.ylim(-1,1)
plt.savefig('fig2.png', dpi = 600)
```
